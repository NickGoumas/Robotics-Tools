```python
# Info on file found below:
# https://sigrok.org/wiki/File_format:Rigol_ROF

#from struct import unpack

filename = '04.ROF'
data_to_save = [['CH1_V', 'CH1_C',
                 'CH2_V', 'CH2_C',
                 'CH3_V', 'CH3_C']]

file_, ext = filename.split(sep='.')

# Open file as a read-only (r) binary (b) file.
with open(filename, 'rb') as f:
    # Make sure there are a correct amount of bytes in the file.
    # 28 bytes for the header data, then three sets of 4 byte pairs
    # for the three channels.
    
    total_bytes = f.seek(0,2)
    header_bytes = 28
    data_bytes = total_bytes - header_bytes
    print('Total Bytes:', total_bytes)
    print('Data Bytes;', data_bytes)
    if data_bytes % (8*3) == 0:
        print('Correct Format')
    else:
        print('Data amounts do not add up')
    
    # Back to beginning.
    f.seek(0)
    
    # Filetype.
    filetype_string = f.read(4)
    print(filetype_string.decode('UTF-8'))
    
    # Metadata header length. Not sure if useful.
    f.seek(6)
    meta_header_len = int.from_bytes(f.read(2), byteorder='little')
    print(meta_header_len)
    
    # Sample period in secodns, little-endian.
    f.seek(16)
    sample_period = int.from_bytes(f.read(4), byteorder='little')
    print(sample_period)
    
    # Number of data points.
    data_points = int.from_bytes(f.read(4), byteorder='little')
    print(data_points)
    
    # Save data for each channel.
    f.seek(28)
    for i in range(data_points):
        line = []
        for num in range(6):
            next_ = int.from_bytes(f.read(4), byteorder='little')
            next_ = float(next_) / 10000
            line.append(next_)
        data_to_save.append(line)

    # print(data_to_save)
    
    f.close()
    
    with open(file_+'.csv', 'w+') as nf:
        nf.write('period_in_seconds,'+str(sample_period)+','+'\n')
        nf.write('total_data_points,'+str(data_points)+','+'\n')
        
        for line in data_to_save:
            for elem in line:
                nf.write(str(elem))
                nf.write(',')
            nf.write('\n')
        nf.close()
    
    
    
```
